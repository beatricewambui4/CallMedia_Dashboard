```python
#drop unecessary columns
df.drop(['BRIEF_DESCRIPTION','DATE_TIME'],axis=1, inplace=True)
```

```python
#handling null values for date column
df = df.dropna(subset=['DATE', 'TIME'])
```

```python
df['AGENCY']=df['AGENCY'].fillna(np.nan)
df['AGENCY_NAME']=df['AGENCY_NAME'].fillna(np.nan)
df['CALL_RESOLUTION']=df['CALL_RESOLUTION'].fillna(np.nan)
```

```python
# Check null values with percentage
null_df = pd.DataFrame({
    'Missing Values': df.isna().sum(),
    'Percentage (%)': (df.isna().sum() / len(df) * 100).round(2)
})
print(null_df)
```

```python
#fix data types of each column 
df.columns=(df.columns
    .str.strip()
    .str.replace(' ','_')
    .str.title()
)
```

```python
#convert data types for the content inside the columns
#string columns
Text_Column=['Agency', 'Agency_Name','Inquiry_Name','Call_Resolution']
df[Text_Column]=df[Text_Column].astype(str)
```

```python
#changing the date and time column names since sql assumes they are keywords they take them as main words
df.rename(columns={
                'Date' : 'Call_Date',
                'Time':'Call_Time'
},
          inplace=True
)
```

```python
from sqlalchemy import text

chunk_size = 5000
total_rows = 0
skipped_rows = 0

for i, chunk in enumerate(pd.read_csv(
    r'C:\ProgramData\MySQL\MySQL Server 8.0\Uploads\improved_cleaned_call_center.csv',
    chunksize=chunk_size
)):
    # ✅ Drop nulls
    chunk = chunk.dropna(subset=['Call_Date', 'Call_Time'])
    
    # ✅ Fix date columns
    chunk['Call_Date'] = pd.to_datetime(chunk['Call_Date'], errors='coerce')
    chunk['Call_Time'] = pd.to_datetime(chunk['Call_Time'], errors='coerce').dt.time
    
    # ✅ Drop rows where date parsing failed
    chunk = chunk.dropna(subset=['Call_Date'])
    
    # ✅ KEY FILTER — Keep 2015 only
    before = len(chunk)
    chunk = chunk[chunk['Call_Date'].dt.year == 2015]
    after = len(chunk)
    skipped_rows += (before - after)
    
    # ✅ Truncate Call_Resolution as safety net
    chunk['Call_Resolution'] = chunk['Call_Resolution'].astype(str).str[:500]
    
    # ✅ Skip empty chunks
    if chunk.empty:
        continue
    
    # ✅ Load into MySQL
    chunk.to_sql(
        name='callcentreimprovedtable',
        con=engine,
        if_exists='append',
        index=False,
        method='multi'
    )
    
    total_rows += len(chunk)
    print(f"✅ Chunk {i+1} loaded — {total_rows} rows so far | Skipped so far: {skipped_rows}")

print(f"\n🎉 Done! Total rows loaded: {total_rows}")
print(f"⚠️ Total rows skipped (non-2015): {skipped_rows}")
```

```python
import pandas as pd

query = """
    SELECT MONTHNAME(Call_Date) AS Month, COUNT(*) AS Total_Calls
    FROM callcentreimprovedtable
    GROUP BY MONTH(Call_Date), MONTHNAME(Call_Date)
    ORDER BY MONTH(Call_Date)
"""
df = pd.read_sql(query, engine)
print(df)
```
